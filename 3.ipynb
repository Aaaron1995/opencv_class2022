```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

vedio = cv2.VideoCapture("h3.mp4")
video2 = cv2.VideoWriter(
    "3.mp4",
    cv2.VideoWriter_fourcc(*'MP4V'),
    30,
    (int(vedio.get(3)),int(vedio.get(4)))
)

while vedio.isOpened() == True:
  ret,m_1 = vedio.read()#會回傳布林值與圖片參數
  if ret == True:
    img = cv2.inRange(m_1, (30, 15, 15), (255, 80, 65))
    img2 = cv2.dilate(img,np.ones((30,17)))
    p, t=cv2.findContours(img2, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
    if len(p) != 0:
      pm = []
      for i in range(0,len(p)):
        pm.append(len(p[i]))
      x, y, w, h = cv2.boundingRect(p[pm.index(max(pm))])
      cv2.rectangle(m_1,(x, y),(x+w, y+h),(0, 0, 255),2)
      video2.write(m_1)
    else:
      video2.write(m_1)
  else:
    break
video2.release()
```
