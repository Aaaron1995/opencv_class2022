```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

img = cv2.imread("2.jpg",1)

img2 = cv2.erode(img,np.ones((10,10)))#侵蝕

img3 = cv2.dilate(img,np.ones((10,10)))#膨脹

cv2_imshow(img2)
cv2_imshow(img3)
```

```python
img = cv2.imread("3.jpg",1)

img2 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, np.ones((7,7)))
cv2_imshow(img2)

img3 = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, np.ones((6,6)))
cv2_imshow(img3)
```

```python
img = cv2.imread("2.jpg",1)

img2 = cv2.inRange(img, (0, 0, 0), (30, 30, 30))#已上下先抓取顏色範圍
cv2_imshow(img2)
```

```python
p, t=cv2.findContours(img2, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
imgk = cv2.drawContours(img, p, -1, (0, 0, 255), 2) 

cv2_imshow(imgk)

print(len(p))
```
